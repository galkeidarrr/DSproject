# Data Sience Project - Task 1  

```python
import pandas as pd
```

```python
#load the csv and show
df = pd.read_csv("hotels_data.csv",parse_dates=['Snapshot Date','Checkin Date'])
df.head()
```

## 1.a

#### adding column - DayDiff

```python
df["DayDiff"] = (df["Checkin Date"] - df["Snapshot Date"]).dt.days
```

```python
df.head()
```

## 1.b

#### adding column - WeekDay

```python
df["WeekDay"] = df["Checkin Date"].dt.day_name()
```

```python
df.head()
```

## 1.c

#### adding column - DiscountDiff

```python
df["DiscountDiff"] = df["Original Price"] - df["Discount Price"]
```

```python
df.head()
```

## 1.d

#### adding column - DiscountPerc

```python
df["DiscountPerc"] = df["DiscountDiff"]/df["Original Price"]*100
```

```python
df.head()
```

#### Save df to csv

```python
df.to_csv("Hotels_data_Changed.csv")
```
